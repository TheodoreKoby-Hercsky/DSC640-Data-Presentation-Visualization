```python
# I will first plo the latitude and longitude values
fig,ax = plt.subplots(figsize = (15,15))
costcos_geocoded.plot(ax = ax)
```

```python
# Next I will create points from the Latitude and Longitude information
from shapely.geometry import Point, Polygon
geometry = [Point(xy) for xy in zip( costcos_geocoded["Longitude"], costcos_geocoded["Latitude"])]
geometry[:3]
```

```python
crs = {'init': 'epsg:4326'}
geopoint_data = gpd.GeoDataFrame(costcos_geocoded, crs = crs, geometry = geometry)
geopoint_data.head()
```

```python
geopoint_data.plot(color="red")
```

```python
# I will create a spatial chart for the geopoint and 
# exclude Alaska and Hawaii
geopoint_data[geopoint_data['State'].isin(
    ['Alaska','Hawaii']) == False].plot(
    figsize=(30,20), color='#3B3C6E');
```
